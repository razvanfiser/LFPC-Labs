```python
#creating the graph
graph = {"S": ["aB", "bB"],
         "B": ["cD"],
         "D": ["dQ", "a"],
         "Q": ["bB", "dQ"],
        }
```

```python
#check_word function checks wether a word can be indeed generated by the grammar
def check_word(word, graph):
    non_termin = "S"
    termin = ""
    path = "S"
    check = True
    
    for letter in word:
        for pair in graph[non_termin]:
            if pair[0] == letter:
                path = path[:-1] + pair
                print(path)
                non_termin = pair[-1]
                termin = pair[0]
                check = True
                break
            else:
                check = False
        if not check:
            return False
    
    if path == word:
        return True
```

```python
#plotting the graph using the graphviz library 
```

```python
import graphviz
```

```python
dot = graphviz.Digraph()
```

```python
#creating the nodes
```

```python
for letter in list(graph):
    dot.node(letter)
```

```python
#creating of list of tuples, each tuples contains (parent, node, weight)
```

```python
nodes_weights = []
for key in list(graph):
    for pair in graph[key]:
        nodes_weights.append((key, pair[-1], pair[0]))
```

```python
#creating the edges
```

```python
for tup in nodes_weights:
    dot.edge(tup[0], tup[1], label=tup[2])
```

```python
#printing the graph to the screen
```
